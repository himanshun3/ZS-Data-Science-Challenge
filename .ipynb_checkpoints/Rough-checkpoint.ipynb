```python
import pandas as pd
import numpy as np
df_train=pd.read_csv('dataset/yds_train2018.csv')
df_test=pd.read_csv('dataset/yds_test2018.csv')
df_sample=pd.read_csv('dataset/sample_submission.csv')
```

```python
df_train=df_train.drop(['Week','Merchant_ID'],axis=1)
```

```python
df_train.head(2)
```

```python
df_test.head(2)
```

```python
total_sales=df_train['Sales']
df_test=df_test.drop('Sales',axis=1)
```

```python
df_test.head(2)
```

```python
df_total=pd.concat([df_train.drop(['Sales'],axis=1),df_test])
```

```python
df_total=pd.get_dummies(df_total,columns=['Country'],drop_first=True)
```

```python
df_total.head(20)
```

```python
df_total=df_total.drop('S_No',axis=1)
```

```python
df_total.head(10)
```

```python
df_total_train=df_total.iloc[:79072]
df_total_test=df_total.iloc[79072:]
df_total_test.shape
```

```python
X=df_total_train.values
y=total_sales
test=df_total_test.values
df_total_train.shape
```

```python
from sklearn.linear_model import Ridge
ridge=Ridge(alpha=0.1,normalize=True)
ridge.fit(X,y)
ridge_pred=ridge.predict(test)
ridge_pred=ridge_pred.astype(int)
df_test['Sales']=ridge_pred
df_test[['S_No','Year','Month','Product_ID','Country','Sales']].to_csv('final_submission_1.csv', index=False)
```
